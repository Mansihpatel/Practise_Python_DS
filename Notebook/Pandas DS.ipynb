# ********* Aggregation and group by

```python
import pandas as pd 
import numpy as np 

car_data = pd.read_csv('Car_sales.csv')
car_data
```

```python
car_data.columns
```

```python
# taking last 15 rows
df = car_data[-10:] #fetching last 10 rows
df = df.reset_index(drop=True)
df
```

```python
# agregate moving average - smoothing - remove deviations

moving_average = df.rolling(window=3,min_periods = 1)# try changing min record

moving_average.aggregate(np.mean)
```

```python
# moving average on specific column
moving_average_sales = df.rolling(window=3,min_periods = 1)# try changing min record

moving_average_sales['Sales_in_thousands'].aggregate(np.mean)
```

```python
# aggregation on multiple columns 
# moving average on specific column
moving_average_salesprice = df.rolling(window=3,min_periods = 1)# try changing min record

moving_average_salesprice['Sales_in_thousands','Price_in_thousands'].aggregate(np.mean)
```

```python
# moving average based on multiple columns and multiple functions
moving_average_salesprice = df.rolling(window=3,min_periods = 1)# try changing min record

moving_average_salesprice['Sales_in_thousands','Price_in_thousands'].aggregate([np.mean,np.min,np.max])
```

```python
# changing data set for group by operation 
df['year'] = df['Latest_Launch'].str.strip().str[-4:]
df['Month']= df['Latest_Launch'].str.strip().str[0:2]
```

```python
df
```

```python
# removing '/' from month column
df.Month=df.Month.str.replace('/','') 
df 
```

```python
# group by year .......... price
df.groupby('year').Price_in_thousands.mean()
```

```python
#lambda on groupby -- on whole data

df.groupby('year').apply(lambda df:df.mean())
```

```python
# group by on multiple columns
df.groupby(['year','Month']).Price_in_thousands.mean()
```

```python
# group by and using multiple aggregation
df.groupby(['year']).Price_in_thousands.aggregate([np.mean,min,max,len,np.std])
```

```python
import pandas as pd 
import numpy as np 

car_data = pd.read_csv('Car_sales.csv')
df1 = car_data.iloc[:5,[0,1,2,14]]
df2 = car_data.iloc[0:5,[3,4,5,14]]
display(df1,df2)
```

```python
df = pd.concat([df1,df2])
df

# basically its stacking data on eac other 
```

```python
# join to combine them on specific column ..e.g 'date'
df_join = df1.join(df2,lsuffix='_LEFT',rsuffix='_RIGHT')
df_join
```

```python
# using join and common column as key/index to make join
df3 = df1.set_index('Latest_Launch').join(df2.set_index('Latest_Launch'))
df3
```

```python
# to preserve the index in the original dataframe then we use below method
df4 = df1.join(df2.set_index('Latest_Launch'),on='Latest_Launch')
df4
```

```python
# merge --- based on any column
df5 = pd.merge(df1,df2,on='Latest_Launch')
df5
```

```python
# joins

dt1 = car_data.iloc[0:6,0:4]
dt2 = car_data.iloc[4:10,0:3]
print(dt1)
print(dt2)
```

```python
# inner join - common values
dt3 = pd.merge(dt1,dt2)
dt3
```

```python
# left join
dt4 = pd.merge(dt1,dt2,how='left',on='Model')
dt4
```

```python
#right join

dt5 = pd.merge(dt1,dt2,how='right',on='Model')
dt5
```

```python
# outer join --- data from both table
dt6 = pd.merge(dt1,dt2,how='outer')
dt6
```

```python
dt6 = pd.merge(dt1,dt2,how='outer',on='Model')
dt6
```
